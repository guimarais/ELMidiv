```python
from ipfnpytools.getsig import getsig
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('helvet2')
```

```python
shotnr = 30554
freq_elm = getsig(shotnr, 'ELM', 'freq_ELM')
f_elm = getsig(shotnr, 'ELM', 'f_ELM')
telm = getsig(shotnr,'ELM', 't_endELM')
```

```python
fmin = 85
fmax = 133
plt.figure(dpi=200)
plt.scatter(freq_elm.time, freq_elm.data, s=12)
plt.scatter(f_elm.time, f_elm.data, s=12)
plt.xlim(3.0,3.5)
plt.ylim(0.0,200)
plt.axhspan(0, fmin, color='k', alpha=0.5)
plt.axhspan(fmax, 1000, color='k', alpha=0.5)
plt.show()
```
